### 5.Extracting only those attributes that are relevant to us such as `Date`, `Time`,`Timestamp`,`Alarm`


```python
insulin_data_frame_filtered = insulin_data_frame[["Date","Time", "Timestamp","Alarm"]]
insulin_data_frame_filtered
```

```python
auto = insulin_data_frame_filtered.sort_values(by='Timestamp', ascending=True).loc[
                                                                insulin_data_frame_filtered['Alarm']=='AUTO MODE ACTIVE PLGM OFF'].iloc[0]['Timestamp']
auto
                                                                                
```

```python
### Creating a data frame which has all auto mode values greater than the auto timestamp.

auto_data_frame = cgm_data_filtered_set.sort_values(by='Timestamp', ascending=True).loc[
                                                                cgm_data_filtered_set['Timestamp'] >=auto]
auto_data_frame
```

*** Creating a Data frame for all manual mode, where all timestamps value are less then auto start timestamp value***

```python

manual_data_frame = cgm_data_filtered_set.sort_values(by='Timestamp', ascending=True).loc[cgm_data_filtered_set['Timestamp'] < auto]
manual_data_frame
```

```python
auto_data_frame_index = auto_data_frame
auto_data_frame_index = auto_data_frame_index.set_index('Timestamp')
auto_data_frame_index
```
